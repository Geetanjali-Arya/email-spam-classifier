```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
df.iloc[0].text
```

```python
def remove_special_chars(text):
    
    result = ""
    
    for i in text:
        if i.isalnum() or i == ' ':
            result = result + i
        else:
            result = result + ""
    return result.strip()
            
```

```python
df['text'] = df['text'].apply(remove_special_chars) #remove spcial characters
```

```python
df.drop([1381,2652],axis=0,inplace=True) #deleting1381 and 2652 no. row because it contains string
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english',max_features=10000) #removing stopwords and take 10000 words
```

```python
X = cv.fit_transform(df['text']).toarray()
```

```python
X.shape
```

```python
X[0]
```

```python
df['spam'] = df['spam'].astype('int')
```

```python
y = df['spam'].values
```

```python
y
```

```python
y.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
```

```python
clf.fit(X_train,y_train) #training model
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
